```python
# 导包
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10, 0.2)
y_sin = np.sin(x)
y_cos = np.cos(x)
fig, ax = plt.subplots(1, 2)
ax[0].plot(x, y_sin)
ax[0].set_xlim(0, 5)  # set_ylim()
ax[0].set_title('sub_title1')
ax[0].set_xlabel('sub_label')
ax[1].plot(x, y_cos)
fig.suptitle('Test')
plt.show()
# plt.setp(fig)  # 调用此方法可以获取 目标可设置行属性的列表
```

```python
data1, data2, data3, data4, data5, data6, data7, data8 = np.random.randn(8, 4)
fig, ax = plt.subplots(1, 4)  # 可以先定义好子类个数。也可以后期自定义子类
ax[0].plot(data1, data2, 'o-')  # 四种线：-、--、:、-.、
ax[1].plot(data3, data4, 'ro--')
ax[2].plot(data5, data6, 'go:')
ax[3].plot(data7, data8, 'yo-.', linewidth=4.0)  # linewidth=2.0  设置先的宽度
fig.set_size_inches(9, 3)
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
plt.figure(1, figsize=(9, 3))
plt.subplot(131)  # 这种类似 plt.subplots(1, 3)
plt.scatter(names, values)
plt.subplot(132)
plt.bar(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.show()
```

```python
x = np.arange(50)
y = x + 10 * np.random.randn(50)
size = np.random.randint(0, 50, 50)
color = np.abs(size) * 100
plt.scatter(x, y, size, color)
plt.ylabel('Scatter Picture')  # plt.axis([0, 100, 0, 100])
plt.text(10, 50, r'hello pyplot')  # 本意是在图中添加文字。但是注释起来有点麻烦
# 注释专用。只需要注意两个点：xy待注释的位置坐标、xytext注释文本的位置坐标
plt.annotate('hello annotate', xy=(10, 20), xytext=(5, 40), arrowprops=dict(arrowstyle='-|>'))  # -|>
plt.show()  # plt.grid(True)  添加网格
```

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]  # 所占比例，平分100份。从x周正方向开始，逆时钟画图
explode = (0, 0.1, 0, 0)  # 爆炸图
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # 这个用了好看点
plt.show()
```
